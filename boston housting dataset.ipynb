```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_boston
boston_ = load_boston()
boston = pd.DataFrame(boston_.data,columns=boston_.feature_names)
boston['MEDV'] = boston_.target
corre_matrix = boston.corr().round(1)
sns.heatmap(corre_matrix,annot=True)
#X = pd.DataFrame(np.c_[boston['LSTAT'], boston['RM']], columns = ['LSTAT','RM'])
```

```python


plt.scatter(boston['LSTAT'],boston['MEDV'], color='r')
plt.xlabel('LSTAT')
plt.ylabel('MEDV')

from sklearn.model_selection import train_test_split
x = boston['LSTAT'].values.reshape(-1,1)
y = boston['MEDV'].values.reshape(-1,1)
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.2, random_state=4)

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error


regressor = LinearRegression().fit(X_train,Y_train)

y_test_predict = regressor.predict(X_test)
rmse = np.sqrt(mean_squared_error(Y_test, y_test_predict))
print('RMSE is {}'.format(rmse))
```

```python
plt.scatter(X_train,Y_train)
plt.plot(X_test,y_test_predict,color = 'r')
print(regressor.coef_,regressor.intercept_)
```
